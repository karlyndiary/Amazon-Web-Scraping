```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
#function to return product title
def get_title(soup):
    try:
        # outer tag object
        title = soup.find("span", attrs = {"id": "productTitle"}).text.strip()
    
        # inner navigatable string object
        #title_value = title.text()
    
        # title as a sting value
        #title_string = title_value.strip()

    except AttributeError:
        title = ""

    return title
```

```python
#function to return product price
def get_price(soup):
    try:
        # outer tag object
        price = soup.find("span", attrs = {"class": "a-price-whole"}).text.strip()

    except AttributeError:
        price = ""

    return price
```

```python
#function to return product rating
def get_rating(soup):
    try:
        # outer tag object
        rating = soup.find("span", attrs={"id":'acrCustomerReviewText'}).text.strip()

    except AttributeError:
        rating = ""

    return rating
```

```python
#function to return product color
def get_color(soup):
    try:
        # outer tag object
        color = soup.find("tr", attrs={"class":'a-spacing-small po-color'}).find("td", attrs={"class":'a-span9'}).find("span", attrs={"class":'a-size-base po-break-word'}).text

    except AttributeError:
        color = ""

    return color
```

```python
#function to return product mrp
def get_mrp(soup):
    try:
        # outer tag object
        mrp = soup.find("span", attrs={"class":'a-price a-text-price'}).find("span", attrs={"class":'a-offscreen'}).text

    except AttributeError:
        mrp = ""

    return mrp
```

```python
#function to return product ram
def get_ram(soup):
    try:
        # outer tag object
        ram = soup.find("tr", attrs={"class":'a-spacing-small po-ram_memory.installed_size'}).find("td", attrs={"class":'a-span9'}).find("span", attrs={"class":'a-size-base po-break-word'}).text

    except AttributeError:
        ram = ""

    return ram
```

```python
#function to return product brand
def get_brand(soup):
    try:
        # outer tag object
        brand = soup.find("tr", attrs={"class":'a-spacing-small po-brand'}).find("td", attrs={"class":'a-span9'}).find("span", attrs={"class":'a-size-base po-break-word'}).text

    except AttributeError:
        brand = ""

    return brand
```

```python
#function to return product storage
def get_storage(soup):
    try:
        # outer tag object
        storage = soup.find("div", attrs={"class":'twisterTextDiv text'}).find("p", attrs={"class":'a-text-left a-size-base'}).text

    except AttributeError:
        storage = ""

    return storage
```

```python
#function to return product discount
def get_discount(soup):
    try:
        # outer tag object
        discount = soup.find("div", attrs={"class":'a-section a-spacing-none aok-align-center aok-relative'}).find("span", attrs={"class":'a-size-large a-color-price savingPriceOverride aok-align-center reinventPriceSavingsPercentageMargin savingsPercentage'}).text

    except AttributeError:
        discount = ""

    return discount
```

```python
#function to return product star
def get_star(soup):
    try:
        # outer tag object
        star = soup.find("a", attrs={"class":'a-popover-trigger a-declarative'}).find("span", attrs={"class":'a-size-base a-color-base'}).text.strip()

    except AttributeError:
        star = ""

    return star
```

```python
#function to return product model
def get_model(soup):
    try:
        # outer tag object
        model = soup.find("tr", attrs={"class":'a-spacing-small po-model_name'}).find("td", attrs={"class":'a-span9'}).find("span", attrs={"class":'a-size-base po-break-word'}).text

    except AttributeError:
        model = ""

    return model
```

```python
#function to return product screen size
def get_screen_size(soup):
    try:
        # outer tag object
        screen_size = soup.find("tr", attrs={"class":'a-spacing-small po-display.size'}).find("td", attrs={"class":'a-span9'}).find("span", attrs={"class":'a-size-base po-break-word'}).text

    except AttributeError:
        screen_size = ""

    return screen_size
```

```python
#function to return product cpu
def get_cpu(soup):
    try:
        # outer tag object
        cpu = soup.find("tr", attrs={"class":'a-spacing-small po-cpu_model.family'}).find("td", attrs={"class":'a-span9'}).find("span", attrs={"class":'a-size-base po-break-word'}).text

    except AttributeError:
        cpu = ""

    return cpu
```

```python
#function to return product os
def get_os(soup):
    try:
        # outer tag object
        os = soup.find("tr", attrs={"class":'a-spacing-small po-operating_system'}).find("td", attrs={"class":'a-span9'}).find("span", attrs={"class":'a-size-base po-break-word'}).text

    except AttributeError:
        os = ""

    return os
```

```python
#function to return product graphic card
def get_graphic_card(soup):
    try:
        # outer tag object
        graphic_card = soup.find("tr", attrs={"class":'a-spacing-small po-graphics_description'}).find("td", attrs={"class":'a-span9'}).find("span", attrs={"class":'a-size-base po-break-word'}).text

    except AttributeError:
        graphic_card = ""

    return graphic_card
```

```python
#function to return product graphic coprocessor
def get_graphic_coprocessor(soup):
    try:
        # outer tag object
        graphic_coprocessor = soup.find("tr", attrs={"class":'a-spacing-small po-graphics_coprocessor'}).find("td", attrs={"class":'a-span9'}).find("span", attrs={"class":'a-size-base po-break-word'}).text

    except AttributeError:
        graphic_coprocessor = ""

    return graphic_coprocessor
```

```python
#function to return product availability
def get_availability(soup):
    try:
        # outer tag object
        availability= soup.find("span", attrs={"class":'a-size-medium a-color-success'}).text.strip()

    except AttributeError:
        availability = ""

    return availability
```

```python
if __name__ == '__main__':
    
    # add user agents
    HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})

    # webpage url
    url = "https://www.amazon.in/s?k=dell+laptop&crid=3IMFI0I7PNCYF&sprefix=dell+lapt%2Caps%2C534&ref=nb_sb_noss_2"

    # https requests
    webpage = requests.get(url, headers=HEADERS)

    # soup object containing all data
    soup = BeautifulSoup(webpage.content, "html.parser")

    # fetch links as list of tag objects
    links = soup.find_all("a", attrs={'class': 'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})

    # store link in list
    link_list = []

    # loop for extracting links from tag objects
    for link in links:
        link_list.append(link.get('href'))

    d = {'title':[], 'price':[], 'rating':[], 'color':[], 'mrp':[], 'ram':[], 'brand':[], 'storage':[], 'discount':[], 'star':[], 
         'model':[], 'screen_size':[], 'cpu':[], 'os':[], 'graphic_card':[], 'graphic_coprocessor':[], 'availability':[]}

    # loop for extracting product details for each time
    for link in link_list:
        new_webpage = requests.get("https://www.amazon.in/" + link, headers=HEADERS)
        
        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        # function calls to display all necessary product information
        d['title'].append(get_title(new_soup))        
        d['price'].append(get_price(new_soup))
        d['rating'].append(get_rating(new_soup))
        d['color'].append(get_color(new_soup))
        d['mrp'].append(get_mrp(new_soup))
        d['ram'].append(get_ram(new_soup))
        d['brand'].append(get_brand(new_soup))
        d['storage'].append(get_storage(new_soup))
        d['discount'].append(get_discount(new_soup))
        d['star'].append(get_star(new_soup))
        d['model'].append(get_model(new_soup))
        d['screen_size'].append(get_screen_size(new_soup))
        d['cpu'].append(get_cpu(new_soup))
        d['os'].append(get_os(new_soup))
        d['graphic_card'].append(get_graphic_card(new_soup))
        d['graphic_coprocessor'].append(get_graphic_coprocessor(new_soup))
        d['availability'].append(get_availability(new_soup))

    amazon_df = pd.DataFrame.from_dict(d)
    amazon_df['title'].replace('', np.nan, inplace=True)
    amazon_df = amazon_df.dropna(subset=['title'])
    amazon_df.to_csv("amazon_data.csv", header=True, index=False)
```
